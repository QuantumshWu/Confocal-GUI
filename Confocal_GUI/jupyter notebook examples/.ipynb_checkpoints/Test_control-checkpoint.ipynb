```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QVBoxLayout, QWidget
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.figure import Figure
import numpy as np

class MyApp(QMainWindow):
    def __init__(self):
        super().__init__()
        
        self.initUI()

    def initUI(self):
        self.setWindowTitle('Matplotlib in PyQt5')
        self.setGeometry(100, 100, 800, 600)

        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        
        layout = QVBoxLayout(central_widget)

        self.canvas = PlotCanvas(self)
        layout.addWidget(self.canvas)

        self.show()

class PlotCanvas(FigureCanvas):
    def __init__(self, parent=None):
        self.figure = Figure()
        self.ax = self.figure.add_subplot(111)
        super().__init__(self.figure)
        self.setParent(parent)
        
        self.init_plot()
        
        self.vline = self.ax.axvline(color='green', linestyle='--', visible=False)
        self.hline = self.ax.axhline(color='green', linestyle='--', visible=False)
        self.text = self.ax.text(0.7, 0.9, '', transform=self.ax.transAxes, visible=False)
        
        self.pressed = False
        self.point = None  # ç”¨äºŽå­˜å‚¨å”¯ä¸€çš„ç‚¹
        
        self.figure.canvas.mpl_connect('button_press_event', self.on_press)
        self.figure.canvas.mpl_connect('button_release_event', self.on_release)
        self.figure.canvas.mpl_connect('motion_notify_event', self.on_motion)
        
    def init_plot(self):
        x = np.linspace(0, 10, 100)
        y = np.sin(x)
        self.ax.plot(x, y)
        self.draw()
        
    def on_press(self, event):
        if event.inaxes:
            if event.button == 1:  # å·¦é”®
                self.pressed = True
                x, y = event.xdata, event.ydata
                self.vline.set_xdata(x)
                self.hline.set_ydata(y)
                self.vline.set_visible(True)
                self.hline.set_visible(True)
                self.text.set_visible(True)
                self.text.set_text(f'x={x:.2f}, y={y:.2f}')
                self.draw()
            elif event.button == 3:  # å³é”®
                x, y = event.xdata, event.ydata
                self.toggle_point(x, y)
                self.draw()

    def on_release(self, event):
        if event.button == 1:  # å·¦é”®
            self.pressed = False

    def on_motion(self, event):
        if self.pressed and event.inaxes:
            x, y = event.xdata, event.ydata
            self.vline.set_xdata(x)
            self.hline.set_ydata(y)
            self.text.set_text(f'x={x:.2f}, y={y:.2f}')
            self.draw()
            print(f'Moved to: ({x}, {y})')

    def toggle_point(self, x, y):
        # å¦‚æžœå·²æœ‰ç‚¹ï¼Œåˆ é™¤å®ƒ
        if self.point is not None:
            self.point.remove()
            self.point = None
        else:
            # å¦‚æžœæ²¡æœ‰ç‚¹ï¼Œæ·»åŠ æ–°ç‚¹
            self.point, = self.ax.plot(x, y, 'ro')  # çº¢è‰²åœ†ç‚¹

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import RectangleSelector
from PyQt5.QtWidgets import QApplication, QMainWindow, QPushButton, QVBoxLayout, QWidget
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.figure import Figure
from matplotlib.animation import FuncAnimation

class LivePlotWidget(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
        self.plotting = False
        self.selector = None

    def initUI(self):
        self.figure = Figure()
        self.canvas = FigureCanvas(self.figure)
        self.ax = self.figure.add_subplot(111)

        self.startButton = QPushButton("Start", self)
        self.startButton.clicked.connect(self.startPlotting)
        self.stopButton = QPushButton("Stop", self)
        self.stopButton.clicked.connect(self.stopPlotting)
        self.selectButton = QPushButton("Select", self)
        self.selectButton.clicked.connect(self.startSelecting)

        layout = QVBoxLayout()
        layout.addWidget(self.canvas)
        layout.addWidget(self.startButton)
        layout.addWidget(self.stopButton)
        layout.addWidget(self.selectButton)
        self.setLayout(layout)

    def startPlotting(self):
        self.plotting = True
        self.x = np.linspace(0, 2 * np.pi, 100)
        self.line, = self.ax.plot(self.x, np.sin(self.x))

        self.ani = FuncAnimation(self.figure, self.updatePlot, blit=True)
        self.canvas.draw()

    def stopPlotting(self):
        self.plotting = False
        if hasattr(self, 'ani'):
            self.ani.event_source.stop()

    def updatePlot(self, frame):
        if self.plotting:
            y = np.sin(self.x + frame / 10.0)
            self.line.set_ydata(y)
            self.ax.draw_artist(self.line)
            self.canvas.blit(self.ax.bbox)
            return self.line,

    def startSelecting(self):
        if self.selector:
            self.selector.set_active(False)
        self.selector = RectangleSelector(self.ax, self.onSelect, interactive=True, useblit=True)
        self.canvas.draw()

    def onSelect(self, eclick, erelease):
        x1, y1 = eclick.xdata, eclick.ydata
        x2, y2 = erelease.xdata, erelease.ydata
        print(f"Rectangle selected from ({x1}, {y1}) to ({x2}, {y2})")

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle("Live Plot with Selection")
        self.setGeometry(100, 100, 800, 600)
        self.plotWidget = LivePlotWidget()
        self.setCentralWidget(self.plotWidget)
        self.show()

if __name__ == '__main__':
    app = QApplication(sys.argv)
    mainWin = MainWindow()
    sys.exit(app.exec_())
```

```python
import numpy as np

from matplotlib.path import Path
from matplotlib.widgets import LassoSelector


class SelectFromCollection:
    """
    Select indices from a matplotlib collection using `LassoSelector`.

    Selected indices are saved in the `ind` attribute. This tool fades out the
    points that are not part of the selection (i.e., reduces their alpha
    values). If your collection has alpha < 1, this tool will permanently
    alter the alpha values.

    Note that this tool selects collection objects based on their *origins*
    (i.e., `offsets`).

    Parameters
    ----------
    ax : `~matplotlib.axes.Axes`
        Axes to interact with.
    collection : `matplotlib.collections.Collection` subclass
        Collection you want to select from.
    alpha_other : 0 <= float <= 1
        To highlight a selection, this tool sets all selected points to an
        alpha value of 1 and non-selected points to *alpha_other*.
    """

    def __init__(self, ax, collection, alpha_other=0.3):
        self.canvas = ax.figure.canvas
        self.collection = collection
        self.alpha_other = alpha_other

        self.xys = collection.get_offsets()
        self.Npts = len(self.xys)

        # Ensure that we have separate colors for each object
        self.fc = collection.get_facecolors()
        if len(self.fc) == 0:
            raise ValueError('Collection must have a facecolor')
        elif len(self.fc) == 1:
            self.fc = np.tile(self.fc, (self.Npts, 1))

        self.lasso = LassoSelector(ax, onselect=self.onselect)
        self.ind = []

    def onselect(self, verts):
        path = Path(verts)
        self.ind = np.nonzero(path.contains_points(self.xys))[0]
        self.fc[:, -1] = self.alpha_other
        self.fc[self.ind, -1] = 1
        self.collection.set_facecolors(self.fc)
        self.canvas.draw_idle()

    def disconnect(self):
        self.lasso.disconnect_events()
        self.fc[:, -1] = 1
        self.collection.set_facecolors(self.fc)
        self.canvas.draw_idle()


if __name__ == '__main__':
    import matplotlib.pyplot as plt

    # Fixing random state for reproducibility
    np.random.seed(19680801)

    data = np.random.rand(100, 2)

    subplot_kw = dict(xlim=(0, 1), ylim=(0, 1), autoscale_on=False)
    fig, ax = plt.subplots(subplot_kw=subplot_kw)

    pts = ax.scatter(data[:, 0], data[:, 1], s=80)
    selector = SelectFromCollection(ax, pts)

    def accept(event):
        if event.key == "enter":
            print("Selected points:")
            print(selector.xys[selector.ind])
            selector.disconnect()
            ax.set_title("")
            fig.canvas.draw()

    fig.canvas.mpl_connect("key_press_event", accept)
    ax.set_title("Press enter to accept selected points.")

    plt.show()
```

```python
import sys
import numpy as np
from PyQt5.QtWidgets import QApplication, QMainWindow, QVBoxLayout, QWidget, QPushButton, QHBoxLayout, QLineEdit
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.figure import Figure

class CrosshairSelector:
    def __init__(self, ax, coord_display):
        self.ax = ax
        self.coord_display = coord_display
        self.cursor_enabled = False
        self.cursor_hline = None
        self.cursor_vline = None
        self.dragging = False

        self.cid_press = ax.figure.canvas.mpl_connect('button_press_event', self.on_press)
        self.cid_motion = ax.figure.canvas.mpl_connect('motion_notify_event', self.on_motion)
        self.cid_release = ax.figure.canvas.mpl_connect('button_release_event', self.on_release)

    def enable_cursor(self):
        self.cursor_enabled = True

    def disable_cursor(self):
        self.cursor_enabled = False
        if self.cursor_hline:
            self.cursor_hline.remove()
            self.cursor_hline = None
        if self.cursor_vline:
            self.cursor_vline.remove()
            self.cursor_vline = None
        self.ax.figure.canvas.draw()
        self.coord_display.clear()

    def on_press(self, event):
        if not self.cursor_enabled or event.inaxes != self.ax:
            return

        # å·¦é”®æŒ‰ä¸‹ï¼Œå¼€å§‹æ‹–åŠ¨
        if event.button == 1:
            self.dragging = True
            if self.cursor_hline and self.cursor_vline:
                self.cursor_hline.remove()
                self.cursor_vline.remove()
            self.cursor_hline = self.ax.axhline(event.ydata, color='red', linewidth=1)
            self.cursor_vline = self.ax.axvline(event.xdata, color='red', linewidth=1)
            self.ax.figure.canvas.draw()

    def on_motion(self, event):
        if not self.dragging or not self.cursor_enabled:
            return
        if event.inaxes != self.ax:
            return
        if self.cursor_hline and self.cursor_vline:
            self.cursor_hline.set_ydata(event.ydata)
            self.cursor_vline.set_xdata(event.xdata)
            self.ax.figure.canvas.draw()
            self.coord_display.setText(f'({event.xdata:.2f}, {event.ydata:.2f})')

    def on_release(self, event):
        if event.button == 1:
            self.dragging = False

class MplCanvas(FigureCanvas):
    def __init__(self, parent=None, width=5, height=4, dpi=100):
        fig = Figure(figsize=(width, height), dpi=dpi)
        self.ax = fig.add_subplot(111)
        super().__init__(fig)
        self.setParent(parent)

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("PyQt5 Matplotlib Example with CrosshairSelector and Coordinate Display")
        self.setGeometry(100, 100, 800, 600)

        self.canvas = MplCanvas(self, width=5, height=4, dpi=100)
        central_widget = QWidget()
        self.setCentralWidget(central_widget)

        self.button = QPushButton('Select', self)
        self.button.setCheckable(True)
        self.button.clicked.connect(self.toggle_cursor)

        self.coord_display = QLineEdit(self)
        self.coord_display.setReadOnly(True)

        layout = QVBoxLayout()
        layout.addWidget(self.canvas)

        button_layout = QHBoxLayout()
        button_layout.addWidget(self.button)
        button_layout.addWidget(self.coord_display)
        layout.addLayout(button_layout)

        central_widget.setLayout(layout)

        self.create_plot()

        self.selector = CrosshairSelector(self.canvas.ax, self.coord_display)

    def create_plot(self):
        x = np.linspace(0, 10, 100)
        y = np.sin(x)
        self.canvas.ax.plot(x, y)
        self.canvas.draw()

    def toggle_cursor(self):
        if self.button.isChecked():
            self.selector.enable_cursor()
        else:
            self.selector.disable_cursor()

if __name__ == '__main__':
    app = QApplication(sys.argv)
    main = MainWindow()
    main.show()
    sys.exit(app.exec_())
```

```python
import sys
import numpy as np
from PyQt5.QtCore import QTimer
from PyQt5.QtWidgets import QApplication, QMainWindow, QVBoxLayout, QWidget, QPushButton, QHBoxLayout, QLineEdit
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.figure import Figure
from matplotlib.widgets import RectangleSelector

class DraggableVLine:
    def __init__(self, line, update_func):
        self.line = line
        self.update_func = update_func
        self.press = None
        self.background = None
        self.cid_press = line.figure.canvas.mpl_connect('button_press_event', self.on_press)
        self.cid_release = line.figure.canvas.mpl_connect('button_release_event', self.on_release)
        self.cid_motion = line.figure.canvas.mpl_connect('motion_notify_event', self.on_motion)

    def on_press(self, event):
        if event.inaxes != self.line.axes:
            return
        contains, _ = self.line.contains(event)
        if not contains:
            return
        self.press = self.line.get_xdata(), event.xdata

    def on_motion(self, event):
        if self.press is None:
            return
        if event.inaxes != self.line.axes:
            return
        xpress, xdata = self.press
        dx = event.xdata - xdata
        self.line.set_xdata(xpress + dx)
        self.update_func()
        self.line.figure.canvas.draw()

    def on_release(self, event):
        self.press = None
        self.line.figure.canvas.draw()

class ImshowZoomPan:
    def __init__(self, ax, cax, hist_ax, coord_display):
        self.ax = ax
        self.cax = cax
        self.hist_ax = hist_ax
        self.coord_display = coord_display
        self.image = None
        self.orig_xlim = None
        self.orig_ylim = None
        self.orig_clim = None
        self.vline_low = None
        self.vline_high = None
        self.cid_scroll = ax.figure.canvas.mpl_connect('scroll_event', self.on_scroll)

    def set_image(self, image):
        self.image = image
        self.orig_xlim = self.ax.get_xlim()
        self.orig_ylim = self.ax.get_ylim()
        self.orig_clim = self.image.get_clim()
        self.update_hist()

    def on_scroll(self, event):
        if event.inaxes == self.ax:
            self.zoom_pan(event)
        elif event.inaxes == self.cax:
            self.change_color_range(event)

    def zoom_pan(self, event):
        xlim = self.ax.get_xlim()
        ylim = self.ax.get_ylim()
        xdata = event.xdata
        ydata = event.ydata

        scale_factor = 1.1 if event.button == 'up' else 0.9

        xlim = [xdata + (x - xdata) * scale_factor for x in xlim]
        ylim = [ydata + (y - ydata) * scale_factor for y in ylim]

        self.ax.set_xlim(xlim)
        self.ax.set_ylim(ylim)
        self.ax.figure.canvas.draw()

    def change_color_range(self, event):
        clim = self.image.get_clim()
        vmin, vmax = clim

        scale_factor = 1.1 if event.button == 'up' else 0.9

        vmin = vmin * scale_factor
        vmax = vmax * scale_factor

        self.image.set_clim(vmin, vmax)
        self.ax.figure.canvas.draw()

    def reset(self):
        self.ax.set_xlim(self.orig_xlim)
        self.ax.set_ylim(self.orig_ylim)
        self.image.set_clim(self.orig_clim)
        if self.vline_low:
            self.vline_low.remove()
        if self.vline_high:
            self.vline_high.remove()
        self.update_hist()
        self.ax.figure.canvas.draw()

    def update_hist(self):
        self.hist_ax.clear()
        data = self.image.get_array().flatten()
        self.hist_ax.hist(data, bins=100, color='gray', alpha=0.75)
        vmin, vmax = self.orig_clim
        self.vline_low = self.hist_ax.axvline(vmin, color='red', linestyle='--', linewidth=1)
        self.vline_high = self.hist_ax.axvline(vmax, color='blue', linestyle='--', linewidth=1)
        self.draggable_low = DraggableVLine(self.vline_low, self.update_clim)
        self.draggable_high = DraggableVLine(self.vline_high, self.update_clim)
        self.hist_ax.figure.canvas.draw()

    def update_clim(self):
        vmin = self.vline_low.get_xdata()[0]
        vmax = self.vline_high.get_xdata()[0]
        self.image.set_clim(vmin, vmax)
        self.ax.figure.canvas.draw()

class MplCanvas(FigureCanvas):
    def __init__(self, parent=None, width=5, height=4, dpi=100):
        fig = Figure(figsize=(width, height), dpi=dpi)
        self.ax = fig.add_subplot(121)
        self.cax = fig.add_axes([0.9, 0.1, 0.03, 0.8])
        self.hist_ax = fig.add_subplot(122)
        super().__init__(fig)
        self.setParent(parent)

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("PyQt5 Matplotlib Example with Imshow Zoom and Pan")
        self.setGeometry(100, 100, 800, 600)

        self.canvas = MplCanvas(self, width=5, height=4, dpi=100)
        central_widget = QWidget()
        self.setCentralWidget(central_widget)

        self.coord_display = QLineEdit(self)
        self.coord_display.setReadOnly(True)

        self.button_reset = QPushButton('Reset', self)
        self.button_reset.clicked.connect(self.reset_view)

        layout = QVBoxLayout()
        layout.addWidget(self.canvas)

        button_layout = QHBoxLayout()
        button_layout.addWidget(self.button_reset)
        button_layout.addWidget(self.coord_display)
        layout.addLayout(button_layout)

        central_widget.setLayout(layout)

        self.create_plot()

        self.zoom_pan = ImshowZoomPan(self.canvas.ax, self.canvas.cax, self.canvas.hist_ax, self.coord_display)
        self.zoom_pan.set_image(self.im)

        self.selector = RectangleSelector(self.canvas.ax, self.on_select,
                                          interactive=True, useblit=True,
                                          props=dict(facecolor='red', edgecolor='black', alpha=0.5, fill=True))

        self.timer = QTimer()
        self.timer.timeout.connect(self.update_plot)
        self.timer.start(1000)  # æ¯ç§’æ›´æ–°ä¸€æ¬¡

    def create_plot(self):
        self.data_buffer = np.zeros((10, 10))
        self.current_index = 0
        self.im = self.canvas.ax.imshow(self.data_buffer, origin='lower')
        self.canvas.figure.colorbar(self.im, cax=self.canvas.cax)
        self.canvas.draw()

    def reset_view(self):
        self.zoom_pan.reset()

    def on_select(self, eclick, erelease):
        x1, y1 = eclick.xdata, eclick.ydata
        x2, y2 = erelease.xdata, erelease.ydata
        self.coord_display.setText(f'({x1:.2f}, {y1:.2f}) to ({x2:.2f}, {y2:.2f})')

    def update_plot(self):
        if self.current_index < self.data_buffer.size:
            self.data_buffer.flat[self.current_index] = 100*np.random.random()
            self.current_index += 1
            self.im.set_data(self.data_buffer)
            self.zoom_pan.update_hist()
            self.canvas.draw()

if __name__ == '__main__':
    app = QApplication(sys.argv)
    main = MainWindow()
    main.show()
    sys.exit(app.exec_())
```

```python
from decimal import Decimal, ROUND_DOWN

def round_to_significant_digits(value, significant_digits):
    if value == 0:
        return 0
    d = Decimal(value)
    return d.scaleb(-d.adjusted() + significant_digits - 1).quantize(Decimal(1), rounding=ROUND_DOWN).scaleb(d.adjusted() - significant_digits + 1)

# ç¤ºä¾‹ä½¿ç”¨
values = [12.34, 0.0223, 12345, 0.000456, 1.234e-5, 1.234000]
significant_digits = 2
rounded_values = [float(round_to_significant_digits(val, significant_digits)) for val in values]

print(rounded_values)  # è¾“å‡ºï¼š[12, 0.022, 12000, 0.00045, 1.2e-05, 1.2]
```

```python
def get_decimal_places(value):
    non_zero = {'1', '2', '3', '4', '5', '6', '7', '8', '9'}
    value_str = str(value)

    if '.' in value_str:
        int_str, float_str = value_str.split('.')

        if any(num in non_zero for num in int_str):
            return 0

        decimal = 0
        for num_str in float_str:
            decimal += 1
            if num_str in non_zero:
                return decimal
    else:
        return 0
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
import time

# ç”Ÿæˆä¸€äº›ç¤ºä¾‹æ•°æ®
data_x = np.linspace(0, 10, 100)
data_y = np.linspace(0, 10, 100)
data_z = np.random.random((100, 100))

# æ¨¡æ‹Ÿå°†é›¶å…ƒç´ ç½®ä¸º0çš„æƒ…å†µ
data_z[::10] = 0

fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(data_z, cmap='inferno')

# åˆ›å»º colorbar
cbar = fig.colorbar(im, ax=ax, orientation='vertical')
cbar.set_label('Intensity')

# ä½¿ç”¨ make_axes_locatable åˆ›å»ºä¸€ä¸ªæ–°çš„è½´ç”¨äºŽç»˜åˆ¶ç›´æ–¹å›¾
divider = make_axes_locatable(ax)
ax_hist = divider.append_axes("left", size="10%", pad=0.4)

# ç›´æ–¹å›¾çš„åˆå§‹æ•°æ®
hist_data = [i for i in data_z.flatten() if i != 0]
n, bins, patches = ax_hist.hist(hist_data, orientation='horizontal', bins=30)

# ç¡®ä¿ç›´æ–¹å›¾ä¸Ž imshow é«˜åº¦ç›¸åŒ
ax_hist.set_ylim(ax.get_ylim())
ax_hist.set_ylabel('Counts')
ax_hist.set_xlabel('Frequency')

# è®¾ç½®èƒŒæ™¯ï¼ˆèƒŒæ™¯åŒ…æ‹¬æ‰€æœ‰é™æ€å…ƒç´ ï¼Œåˆå§‹ç»˜å›¾ï¼‰
fig.canvas.draw()
bg = fig.canvas.copy_from_bbox(fig.bbox)

def update_data():
    """ç”Ÿæˆæ–°çš„éšæœºæ•°æ®"""
    #new_data_z = np.random.poisson(100, (100, 100))
    new_data_z = 100*np.random.random((100, 100))
    #new_data_z[::10] = 0  # æ¨¡æ‹Ÿå°†é›¶å…ƒç´ ç½®ä¸º0çš„æƒ…å†µ
    return new_data_z

def update_plot(new_data_z):
    """æ›´æ–°imshowå’Œç›´æ–¹å›¾"""
    # æ¢å¤èƒŒæ™¯
    fig.canvas.restore_region(bg)
    
    # æ›´æ–° imshow æ•°æ®
    im.set_array(new_data_z)
    
    # æ›´æ–°ç›´æ–¹å›¾æ•°æ®
    new_hist_data = [i for i in new_data_z.flatten() if i != 0]
    n, _ = np.histogram(new_hist_data, bins=bins)
    #ax_hist.set_ylim(np.min(new_data_z), np.max(new_data_z))
    
    for count, patch in zip(n, patches):
        patch.set_width(count)
    
    # é‡æ–°ç»˜åˆ¶æ›´æ–°çš„éƒ¨åˆ†
    ax.draw_artist(im)
    for patch in patches:
        ax_hist.draw_artist(patch)
    
    # æ›´æ–°å±å¹•æ˜¾ç¤º
    fig.canvas.blit(fig.bbox)
    fig.canvas.flush_events()

# å¼€å¯äº¤äº’æ¨¡å¼
plt.ion()

# æ¨¡æ‹Ÿå®žæ—¶æ›´æ–°è¿‡ç¨‹
for _ in range(100):
    new_data_z = update_data()
    update_plot(new_data_z)
    time.sleep(0.1)  # æ¨¡æ‹Ÿå®žæ—¶æ•°æ®æ›´æ–°çš„å»¶è¿Ÿ

plt.ioff()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
import time

class LivePlot:
    def __init__(self, data_x, data_y, data_z, zlabel):
        self.data_x = data_x
        self.data_y = data_y
        self.data_z = data_z
        self.zlabel = zlabel
        self.max_n = 300
        self.init_plot()
        self.vmax = 0
        self.vmin = 0
        #self.max_n = 5

    def init_plot(self):
        self.fig, self.ax = plt.subplots(figsize=(8, 6))
        cmap_ = plt.cm.get_cmap('inferno')
        cmap = cmap_.copy()
        cmap.set_under(cmap_(0))
        
        half_step_x = 0.5 * (self.data_x[-1] - self.data_x[0]) / len(self.data_x)
        half_step_y = 0.5 * (self.data_y[-1] - self.data_y[0]) / len(self.data_y)
        extents = [self.data_x[0] - half_step_x, self.data_x[-1] + half_step_x,
                   self.data_y[-1] + half_step_y, self.data_y[0] - half_step_y]

        self.im = self.ax.imshow(self.data_z, animated=True, alpha=0.7, cmap=cmap, extent=extents)

        divider = make_axes_locatable(self.ax)
        self.axright = divider.append_axes("top", size="10%", pad=0.1)

        if np.max(self.data_z.flatten()) == 0:
            hist_data = [0, 0]
        else:
            hist_data = [i for i in self.data_z.flatten() if i != 0]

        self.n, self.bins, self.patches = self.axright.hist(hist_data, orientation='vertical', bins=100, color='grey')
        self.axright.set_ylim(0, self.max_n)
        self.axright.set_xlim(np.min(hist_data), np.max(hist_data))

        self.cax = divider.append_axes("right", size="5%", pad=0.1)
        cbar = self.fig.colorbar(self.im, cax=self.cax)
        cbar.set_label(self.zlabel)

        #self.axright.set_ylim(self.ax.get_ylim())
        self.axright.set_ylabel('Counts')
        self.axright.set_xlabel('Frequency')

        self.fig.tight_layout()
        self.fig.canvas.draw()
        self.bg = self.fig.canvas.copy_from_bbox(self.fig.bbox)

    def update_plot(self, new_data_z):
        
        if np.max(self.n) > self.max_n:
            self.max_n = np.max(self.n) + 10
            self.axright.set_ylim(0, self.max_n)
            self.axright.set_xlim(np.min(new_data_z.flatten()), np.max(new_data_z.flatten())+10)
            self.fig.canvas.draw()
            self.bg = self.fig.canvas.copy_from_bbox(self.fig.bbox)
            
            
        self.fig.canvas.restore_region(self.bg)

        self.im.set_array(new_data_z)

        if np.max(new_data_z.flatten()) == 0:
            hist_data = [0, 0]
        else:
            hist_data = [i for i in new_data_z.flatten() if i != 0]

        self.n, _ = np.histogram(hist_data, bins=100)

        for count, patch in zip(self.n, self.patches):
            patch.set_height(count)
            patch.set_width(0.5)
        
        self.axright.draw_artist(self.axright.patch)
        for patch in self.patches:
            self.axright.draw_artist(patch)

        self.fig.canvas.blit(self.fig.bbox)
        self.fig.canvas.flush_events()

# ç¤ºä¾‹æ•°æ®
data_x = np.linspace(0, 10, 100)
data_y = np.linspace(0, 10, 100)
data_z = np.random.poisson(100, (100, 100))

plot = LivePlot(data_x, data_y, data_z, 'Z Label')

def update_data():
    """ç”Ÿæˆæ–°çš„éšæœºæ•°æ®"""
    new_data_z = np.random.poisson(100*np.random.random(), (100, 100))
    new_data_z[::10] = 0  # æ¨¡æ‹Ÿå°†é›¶å…ƒç´ ç½®ä¸º0çš„æƒ…å†µ
    return new_data_z

plt.ion()

# æ¨¡æ‹Ÿå®žæ—¶æ›´æ–°è¿‡ç¨‹
for _ in range(100):
    new_data_z = update_data()
    plot.update_plot(new_data_z)
    time.sleep(0.1)  # æ¨¡æ‹Ÿå®žæ—¶æ•°æ®æ›´æ–°çš„å»¶è¿Ÿ

plt.ioff()
plt.show()
```
